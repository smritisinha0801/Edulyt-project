```python
district_df["A3"] = pd.to_numeric(district_df["A3"], errors='coerce')

# Now safely get top and bottom 5
top5_pop = district_df.nlargest(5, "A3")["A1"].tolist()
bottom5_pop = district_df.nsmallest(5, "A3")["A1"].tolist()
# Filter transactions from these districts
top5_txn = filtered_txn[filtered_txn["district_id"].isin(top5_pop)]
bottom5_txn = filtered_txn[filtered_txn["district_id"].isin(bottom5_pop)]

# Aggregate Credit and Debit separately
def txn_summary(df, label):
    return df.groupby("type")["amount"].sum().reset_index().assign(DistrictGroup=label)

top5_summary = txn_summary(top5_txn, "Top 5 Populated")
bottom5_summary = txn_summary(bottom5_txn, "Bottom 5 Populated")

# Combine both summaries
final_summary = pd.concat([top5_summary, bottom5_summary], ignore_index=True)

# Show result
print(final_summary)
```

```python
#Number of cards issued in district where average salary is more than 9000, is it a good strategy?
district_df["A6"] = pd.to_numeric(district_df["A6"], errors='coerce')
high_salary_districts = district_df[district_df["A6"] > 9000]["A1"].unique()

high_salary_accounts = new_account_df[
    new_account_df["district_id"].isin(high_salary_districts)
]

linked_dispositions = new_disposition_df[
    new_disposition_df["account_id"].isin(high_salary_accounts["account_id"])
]

cards_in_high_salary = new_card_df[
    new_card_df["disp_id"].isin(linked_dispositions["disp_id"])
]

transactions_in_high_salary = new_transaction_df[
    new_transaction_df["account_id"].isin(high_salary_accounts["account_id"])
]

total_credit = transactions_in_high_salary[
    transactions_in_high_salary["type"].str.lower() == "credit"
]["amount"].sum()

total_debit = transactions_in_high_salary[
    transactions_in_high_salary["type"].str.lower() == "debit"
]["amount"].sum()

print("Number of cards issued in high-salary districts:", len(cards_in_high_salary))
print("Total credit transactions in those districts:", total_credit)
print("Total debit transactions in those districts:", total_debit)
```

```python
# Are we providing loans to members belonging to district where committed crimes are more than 6000 for code 95, if yes then provide the number of loans per district?
district_df["A11"] = pd.to_numeric(district_df["A11"], errors="coerce")
high_crime_districts = district_df[district_df["A11"] > 6000]["A1"].unique()

accounts_in_high_crime = new_account_df[
    new_account_df["district_id"].isin(high_crime_districts)
]

loans_in_high_crime = loan_df[
    loan_df["account_id"].isin(accounts_in_high_crime["account_id"])
]

loans_with_districts = loans_in_high_crime.merge(
    new_account_df[["account_id", "district_id"]],
    on="account_id",
    how="left"
)
loan_counts_by_district = loans_with_districts["district_id"].value_counts().reset_index()
loan_counts_by_district.columns = ["district_id", "number_of_loans"]

print("Number of loans in high-crime districts (crime code 95 > 6000):")
print(loan_counts_by_district)
```

```python
# How much money was collected from other banks for customer belongs to districts where unemployment rate for any year is greater than 2%.
district_df["A3"] = pd.to_numeric(district_df["A3"], errors="coerce")
district_df["A4"] = pd.to_numeric(district_df["A4"], errors="coerce")

high_unemployment_districts = district_df[
    (district_df["A3"] > 2) | (district_df["A4"] > 2)
]["A1"].unique()

high_unemp_accounts = new_account_df[
    new_account_df["district_id"].isin(high_unemployment_districts)
]["account_id"].unique()

filtered_transactions = new_transaction_df[
    (new_transaction_df["account_id"].isin(high_unemp_accounts)) &
    (new_transaction_df["type"].str.lower() == "credit") &
    (new_transaction_df["operation"].str.lower().str.contains("collection from another bank"))
]
filtered_transactions["amount"] = pd.to_numeric(filtered_transactions["amount"], errors="coerce")
total_collected = filtered_transactions["amount"].sum()

print(f"total money collected from other banks (for high-unemployment districts): {total_collected:.2f}")
```

```python
# Create profile of customers in accordance of districts where max money is being paid to
#a. Insurance.
#b. Household
#c. Leasing
#d. Loan
new_transaction_df["k_symbol"] = new_transaction_df["k_symbol"].str.lower().str.strip()
new_transaction_df["amount"] = pd.to_numeric(new_transaction_df["amount"], errors="coerce")

transactions_with_district = new_transaction_df.merge(
    new_account_df[["account_id", "district_id"]],
    on="account_id",
    how="left"
)

categories = {
    "insurance payment": "Insurance",
    "household": "Household",
    "leasing": "Leasing",
    "loan payment": "Loan"
}

profile_summary = []

for k_symbol_value, label in categories.items():
    filtered = transactions_with_district[
        transactions_with_district["k_symbol"] == k_symbol_value
    ]

    if filtered.empty:
        print(f" No transactions found for: {label}")
        continue

    grouped = filtered.groupby("district_id")["amount"].sum().reset_index()

    if grouped.empty:
        print(f" No amounts to sum for: {label}")
        continue

    max_row = grouped.loc[grouped["amount"].idxmax()]
    profile_summary.append({
        "category": label,
        "district_id": int(max_row["district_id"]),
        "amount_paid": float(max_row["amount"])
    })

profile_df = pd.DataFrame(profile_summary)

profile_df = profile_df.merge(
    district_df[["A1", "A2"]],
    left_on="district_id",
    right_on="A1",
    how="left"
).drop(columns="A1").rename(columns={"A2": "district_name"})

print("District profile where maximum amount paid per category:")
print(profile_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

plt.figure(figsize=(6, 6))
sns.barplot(
    data=profile_df,
    x="category",
    y="amount_paid",
    hue="district_name",
    dodge=False,
    palette="Set2"
)
plt.title("Max Money Paid by District per Transaction Category", fontsize=14)
plt.xlabel("Transaction Category")
plt.ylabel("Amount Paid")
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
# # Create profile of customers in accordance of districts for the status of loan payment, there will be 4 categories.
loan_df["amount"] = pd.to_numeric(loan_df["amount"], errors="coerce")
loan_df["payments"] = pd.to_numeric(loan_df["payments"], errors="coerce")

loan_df["repayment_ratio"] = loan_df["payments"] / loan_df["amount"]

def categorize_repayment(ratio):
    if ratio >= 0.95:
        return "Fully Paid"
    elif ratio >= 0.75:
        return "Mostly Paid"
    elif ratio >= 0.5:
        return "Partially Paid"
    else:
        return "Poorly Paid"

loan_df["repayment_status"] = loan_df["repayment_ratio"].apply(categorize_repayment)

loan_with_district = loan_df.merge(
    new_account_df[["account_id", "district_id"]],
    on="account_id",
    how="left"
)
status_profile = loan_with_district.groupby(["district_id", "repayment_status"]).size().reset_index(name="count")
status_profile = status_profile.merge(
    district_df[["A1", "A2"]],
    left_on="district_id",
    right_on="A1",
    how="left"
).drop(columns="A1").rename(columns={"A2": "district_name"})

print("Customer loan repayment profile by district:")
print(status_profile)
```

```python
# Relate the output of above with district conditions like Crime, Unemployment Rate and Average Salary.
district_factors = district_df[["A1", "A11", "A3", "A4", "A6"]] \
    .rename(columns={
        "A1": "district_id",
        "A11": "crime_code_95",
        "A3": "unemp_95",
        "A4": "unemp_96",
        "A6": "avg_salary"
    })

status_with_factors = status_profile.merge(district_factors, on="district_id", how="left")

print("Loan repayment status with district factors:")
print(status_with_factors.head())
```

```python
# Can we say customers from Bohemia are the ones having more male customers possessing Gold cards in comparison of Moravia?
new_card_df["type"] = new_card_df["type"].str.lower().str.strip()
gold_cards = new_card_df[new_card_df["type"] == "GOLD"]  # Note: 'gold' in lowercase

gold_disp = gold_cards.merge(
    new_disposition_df[["disp_id", "client_id", "account_id"]],
    on="disp_id",
    how="left"
)
gold_disp_client = gold_disp.merge(
    new_client_df[["client_id", "gender"]],
    on="client_id",
    how="left"
)
gold_male = gold_disp_client[gold_disp_client["gender"].str.lower() == "MALE"]

gold_male = gold_male.merge(
    new_account_df[["account_id", "district_id"]],
    on="account_id",
    how="left"
)
gold_male = gold_male.merge(
    district_df[["A2", "A3"]],
    left_on="district_id",
    right_on="A2",
    how="left"
).rename(columns={"A3": "region"})

gold_male["region"] = gold_male["region"].astype(str).str.lower().str.strip()

bohemia_moravia_regions = [
    "central bohemia", "south bohemia", "west bohemia",
    "north bohemia", "east bohemia", "south moravia", "north moravia"
]
gold_male = gold_male[gold_male["region"].isin(bohemia_moravia_regions)]

region_counts = gold_male["region"].value_counts().reset_index()
region_counts.columns = ["region", "male_gold_card_count"]

print("Male customers with Gold cards by region:")
print(region_counts)
```

```python
# Can we say that customers having Classic and Junior card are the ones who are more in debt.
new_card_df["type"] = new_card_df["type"].str.lower().str.strip()

cj_cards = new_card_df[new_card_df["type"].isin(["classic", "junior"])]

cj_disp = cj_cards.merge(
    new_disposition_df[["disp_id", "account_id", "client_id"]],
    on="disp_id",
    how="left"
)

cj_with_loans = cj_disp.merge(
    loan_df[["account_id", "amount", "payments"]],
    on="account_id",
    how="inner"
)
cj_with_loans["amount"] = pd.to_numeric(cj_with_loans["amount"], errors="coerce")
cj_with_loans["payments"] = pd.to_numeric(cj_with_loans["payments"], errors="coerce")
cj_with_loans["outstanding_debt"] = cj_with_loans["amount"] - cj_with_loans["payments"]

debt_summary = cj_with_loans.groupby("type")["outstanding_debt"].sum().reset_index()

debt_summary = debt_summary.sort_values(by="outstanding_debt", ascending=False)

print("Debt profile of Classic and Junior card holders:")
print(debt_summary)
```

```python
import numpy as np
from datetime import datetime

# How will you analyze the performance of Mid age vs adults in terms of loan repayments.
loan_df["amount"] = pd.to_numeric(loan_df["amount"], errors="coerce")
loan_df["payments"] = pd.to_numeric(loan_df["payments"], errors="coerce")
loan_df["repayment_ratio"] = loan_df["payments"] / loan_df["amount"]

loan_clients = loan_df.merge(
    new_disposition_df[["account_id", "client_id"]],
    on="account_id",
    how="left"
)
client_info = new_client_df.copy()
client_info["birth_number"] = client_info["birth_number"].astype(str).str.zfill(6)

client_info["year"] = client_info["birth_number"].str[:2].astype(int)
client_info["month"] = client_info["birth_number"].str[2:4].astype(int)
client_info["day"] = client_info["birth_number"].str[4:6].astype(int)

client_info["month"] = client_info["month"].apply(lambda x: x - 50 if x > 50 else x)

client_info["year"] = client_info["year"].apply(lambda y: 1900 + y if y > 24 else 2000 + y)

client_info["birth_date"] = pd.to_datetime(dict(year=client_info["year"], month=client_info["month"], day=client_info["day"]), errors='coerce')

today = pd.to_datetime("2025-01-01")
client_info["age"] = (today - client_info["birth_date"]).dt.days // 365

loan_with_age = loan_clients.merge(
    client_info[["client_id", "age"]],
    on="client_id",
    how="left"
)
def categorize_age(age):
    if 31 <= age <= 45:
        return "Mid-age"
    elif age > 45:
        return "Adult"
    else:
        return "Other"

loan_with_age["age_group"] = loan_with_age["age"].apply(categorize_age)

valid_age_data = loan_with_age[loan_with_age["age_group"].isin(["Mid-age", "Adult"])]

repayment_analysis = valid_age_data.groupby("age_group").agg({
    "repayment_ratio": ["mean", lambda x: (x >= 0.95).mean()],
    "account_id": "count"
}).reset_index()

repayment_analysis.columns = ["age_group", "avg_repayment_ratio", "percent_fully_repaid", "loan_count"]

print("Loan Repayment Performance by Age Group:")
print(repayment_analysis)
```
